## Marketaux

For more information, check out their [pricing plans](https://www.marketaux.com/pricing) and [documentation](https://www.marketaux.com/documentation).

Each request returns information for 3 articles and you're limited to 100 requests per day. Their API goes as far back as January 2021 as of the time of writing. 


```python
# import statements
import pandas as pd
import json
import os
import requests
from dotenv import load_dotenv
```

```python
# set url and response, update API token
marketaux_url = "https://api.marketaux.com/v1/news/all?api_token=insert_api_token_here&symbols=TSLA&filter_entities=true&language=en&published_after=2021-01-01&published_before=2023-01-31"
marketaux_response = requests.get(marketaux_url).json()
```

```python
# print(json.dumps(marketaux_response, indent=5, sort_keys=True))

# pull the data into a dataframe
df = pd.DataFrame(data = marketaux_response["data"])
df.sort_values("published_at", ascending = True)
```

### Request parameters

You can specify lots of information in the request parameters, such as: symbols, exchanges, entity_types, industries, countries, sentiment_gte, sentiment_lte, min_match_score, filter_entities, must_have_entities, group_similar, search, domains, exclude_domains, source_ids, exclude_source_ids, language, published_before, publisehd_after, published_on, sort, sort_order, limit, and page.

Each article has an associated sentiment_score (between -1 and 1), with negative numbers indicating negative sentiment, 0 indicating neutral sentiment, and positive numbers indicating postive sentiment.

With the free version, each request returns a maxium of 3 articles. Specify pages in the request parameters to continue getting the additional information. Just add "&page=2" to grab the next 3 results, etc. 

Check out the documentation for more details on the API request parameters.
