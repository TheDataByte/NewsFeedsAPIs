## Google News API

For more information, check out their [pricing plans](https://newsapi.org/pricing) and [documentation](https://newsapi.org/docs/endpoints).

Each request returns information for up to 100 articles and you're limited to 100 requests per day. Their API pulls information as far back as one month ago with the free plan.

```python
# import statements
import pandas as pd
import json
import os
import requests
from dotenv import load_dotenv
```

### v2/everything

```python
newsapi_url = "https://newsapi.org/v2/everything?q=Micron&from=2022-06-15&sortBy=popularity&apiKey=insert_api_key_here"
newsapi_response = requests.get(newsapi_url).json()
```

```python
# check out the data
# print(json.dumps(newsapi_response, indent=5, sort_keys=True))
```

```python
df = pd.DataFrame(data = newsapi_response["articles"])

#parse out information from original 'source' column into separate 'id' and 'name' columns
#https://stackoverflow.com/questions/38231591/split-explode-a-column-of-dictionaries-into-separate-columns-with-pandas
df = pd.concat([df.drop(['source'], axis=1), df['source'].apply(pd.Series)], axis=1)
df.sort_values("publishedAt", ascending = True, inplace = False).head()
```

### Request parameters

You can specify lots of information in the request parameters, such as: q (keywords or phrases), serachIn, sources, domains, excludeDomains, from, to, language, sortBy, pageSize (number of results to return per page), and page.

With the free version, each request returns a maximum of 100 articles. Specify pages in the request parameters to continue getting the additional information (for example page=2 would return articles 101-200, etc).  

Check out the documentation for more details on the API request parameters.
