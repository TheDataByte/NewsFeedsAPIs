### Request parameters

You can specify lots of information in the request parameters, such as: country, category, language, domain, from_date, to_date, q (keywords or phrases), qInTitle (keywords or phrases in title only), and page. 

With the free version, each request returns a maxium of 10 articles. Specify pages in the request parameters to continue getting the additional information. Just add "&page=2" to grab the next 10 results, etc. 

At the time of writing this API returns the latest news data and stretches as far back as 2 days prior to the request.
